Merkezi limit teorisine göre örneklem sayısı n = 30 a yaklaştıkça normal dağılıma benzediğini söylemiştik. Yani eşik değer 30 olarak düşünülebilir serbestlik derecesi için az not tuttuk çünkü yemeğin içindeki baharat gibi bir öz olarak konuların içersine serpilmiş olarak anlaşılan bir konu. 30 un nedeni 30 a kadar eklenen her bir gözlem popülasyonu anlama açısından pozitif katkı yapıyor ama bu katkı 30.örneklemden sonra çok ciddi azalıyor, yani 30 bizim için önemli bir eşik o noktadan sonra dağılım normal dağılıma çok benziyor. Normal hayatta zor olan işlerden mesela bir kanser ilacının denenmesi diyelim ki bu gibi işlerde 30 örneklem demek 30 tane kanser adam üstünde ilacın denenmesi sonuç için geçecek zamanda bu kişilerin ölmesi gibi durumlar vs. gözlemin başarılı bir şekilde tamamlanmasını çok zorlaştırıyor. İşte bu noktada yani n<30 olduğu durumlarda normal dağılım olmayacağı için nasıl hareket etmemiz gerek dediğimiz anda Student-t veya t dist. unu kullanacağız. Tekrar arada serbestlik derecesi dememin nedeni onu burada kullanacağız. Normal dağılımda kuyruk yapısı standart sapmaya göre değişiyordu burada da serbestlik derecesine göre değişecek.
```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0) # ilkokul numaram ile tohumluyorum generatoru
veri1t = stats.t.rvs(loc=0,df=1,size=15)
veri2t = stats.t.rvs(loc=0,df=2,size=15)
veri3t = stats.t.rvs(loc=0,df=4,size=15)


sns.displot(veri1t,kind='kde')
sns.displot(veri2t,kind='kde')
sns.displot(veri3t,kind='kde')
plt.xlim(-5,5)
```
