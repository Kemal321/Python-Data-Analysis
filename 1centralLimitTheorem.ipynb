```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
yas = np.random.uniform(low=18, high=75, size = 40000)

#plt.hist(yas,color = "Red", edgecolor = "Black",rwidth=0.5)

sample = random.choices(yas,k=500) # k tane örneklemi yaştan çek demek bu choice yani seç
# tabi bu doğru olan değil merkezi limit teoremi çekilen örneklemlerin ortalamalarının standardt mean dist
# e yakınsadığını söylüyordu. List comprehensizve ile hemen çekelim ortalamları
sampleMeans = [np.mean(random.choices(yas,k=30)) for _ in range(1000)]
plt.hist(sample,color = "Red", edgecolor = "Black",rwidth=0.5)
plt.hist(sampleMeans)
plt.show()

# Örnekte görüldüğü gibi aslında standart mean dist e yakınsamış bir dağılım görüyoruz normalde düz dağılıma
# sahip bir popülasyonumuz vardı
```

```python
import seaborn as sns # genel import şeklide bu şekildedir. 
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(35,1,1000)
plt.hist(x)
plt.show
# bu matplotlib kütüphanesinin verdiği çıktı
```

```python
sns.histplot(x)
plt.show()
# çıktı da da görüleceği üzere dağılım grafiği daha iyi bir grafik gelecektir. aynı zamanda direkt olarak dağılım grafiği de mevcut
sns.displot(x) # şuan için aynı görüntü gibi görünse de ekstra olarak kullanabileceğimiz tooları mevcut bunları göreceğiz. mesela
# dağılımın çizgisini görebiliriz.
sns.displot(x,kde = True) # 3. grafiğe bakacak olursak dağılımın çizgisi ile beraber bize çıktı olarak vermiş bulunmakta.
plt.show()
# sadece yoğunluk grafiğini istersek kdeplot() fonksiyonu direktmen bize dağılım çizgi grafiğini verecektir. Bu şekilde ekstra diğer özellik
# lerini de kullanacağım ve öğreneceğim notlarımı da yüklemiş olacağım.
```

<h1><center> Z-Score Örnek</center></h1>

```python
np.random.seed(3) # bu random sayı üreticinin sabit değerleri üretmesi için o üretim yolunu tekrar kullanarak 
# aynı sayıları üretmesini istediğimiz için. 
veri = np.random.normal(35,2,40000) 
sns.displot(veri)
plt.title("Veri Dağılım Grafiği", loc="right",c="red",fontsize=15)
plt.ylabel("Frekans")
plt.xlabel("Veriler")
plt.axvline(x=np.mean(veri),linestyle="--",linewidth=2.5,label="Ortalama",c="red")
plt.show()

# burada şimdi standardazisyon yaparak görelim formülümüz z = x - mean / std idi
veriStandard = (veri-np.mean(veri))/np.std(veri)
sns.displot(veriStandard)
plt.title("Veri Dağılım Grafiği", loc="right",c="red",fontsize=15)
plt.ylabel("Frekans")
plt.xlabel("Veriler")
plt.axvline(x=np.mean(veriStandard),linestyle="--",linewidth=2.5,label="Ortalama",c="red")
plt.show()
# gördüğümüz gibi burada artık mean değerimiz 0 civarında 
print(np.mean(veriStandard))
```
